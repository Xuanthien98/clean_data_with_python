```python
import pandas as pd
import os
```

```python
# Đọc file dữ liệu
df_original = pd.read_csv("ESG_Series_original.csv")
```

```python
# Kiểm tra 5 dòng đầu tiên của dataset
df_original.head()
```

```python
# Xóa các kí tự thừa [YR....] ở các cột năm
df_original.columns = df_original.columns.str.replace(r'\s\[YR\d+\]', '', regex=True)
```

```python
df_original
```

```python
# Xóa cột không cân thiết: Country Code và Series Code
df_original = df_original.drop(['Country Code', 'Series Code'], axis=1)
```

```python
df_original
```

```python
# Nhận thấy một số dòng cuối bị thừa nên kiểm tra thử 10 dòng cuối
df_original.tail(10)
```

```python
# Loại bỏ 5 dòng thừa cuối (dòng 13632 - 13636)
df_original = df_original.iloc[:-5]
df_original

# Thay giá trị '..' thành null
df_original.replace("..", pd.NA, inplace=True)
```

```python
df_original
```

```python
# Chỉ lấy dữ liệu từ 2016 - 2020
cols_to_keep =  ['Country Name',  'Series Name'] + [str(year) for year in range(2016, 2021)]
df_series = df_original[cols_to_keep]
df_series
```

```python
# Transpose cột Series Name với cột các năm 
df_melted = df_series.melt(id_vars=['Country Name',  'Series Name'], var_name='Year', value_name='Value')
pivoted_df = df_melted.pivot(index=['Country Name',  'Year'], columns='Series Name', values='Value')
pivoted_df = pivoted_df.reset_index().rename_axis(None, axis=1)
pivoted_df
```

```python
# Tính % null trong mỗi cột để xem những tiêu chí nào ít null nhất
null_percents = pivoted_df.isnull().mean() * 100
null_percents.sort_values(ascending=True).head(35)
```

```python
# Lựa các tiêu chí cần thiết và ít dữ liệu null cho E,S,G
df = pivoted_df.loc[:, ['Country Name','Year',
                        'CO2 emissions (metric tons per capita)',
                        'Agriculture, forestry, and fishing, value added (% of GDP)', 
                        'Energy intensity level of primary energy (MJ/$2017 PPP GDP)', 
                        'Land Surface Temperature',
                        'Nitrous oxide emissions (metric tons of CO2 equivalent per capita)',
                        'Methane emissions (metric tons of CO2 equivalent per capita)',
                        'Population ages 65 and above (% of total population)', 
                        'Unemployment, total (% of total labor force) (modeled ILO estimate)',
                        'Mortality rate, under-5 (per 1,000 live births)', 
                        'Life expectancy at birth, total (years)', 'Fertility rate, total (births per woman)', 
                        'Government expenditure on education, total (% of government expenditure)',
                        'Control of Corruption: Estimate', 'Government Effectiveness: Estimate', 
                        'GDP growth (annual %)',
                        'Regulatory Quality: Estimate','Rule of Law: Estimate',
                        'Voice and Accountability: Estimate',
                        'Political Stability and Absence of Violence/Terrorism: Estimate', 
                        'School enrollment, primary and secondary (gross), gender parity index (GPI)'
                        ]
                    ]
df
```

```python
# Nhận thấy datatype của các cột tiêu chí chưa chính xác
df.info()
```

```python
# Thay đổi datatype của các cột tiêu chí thành numeric
df.iloc[:, 2:] = df.iloc[:, 2:].apply(pd.to_numeric)
df.info()
```

```python
# Fill những giá trị null của từng cột = giá trị trung bình của cột đó

# Chỉ fill các cột tiêu chí (trừ 2 cột đầu: Country Name và Year)
cols_to_fillna = df.columns[2:] 
df[cols_to_fillna] = df[cols_to_fillna].fillna(df[cols_to_fillna].mean()) 
df
```

```python
# Kiểm tra dữ liệu đã hết null chưa
df.isnull().sum()
```

```python
# Xuất file csv
df.to_csv('ESG_cleaned.csv', index=False)
```
